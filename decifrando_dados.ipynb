```python

# DADOS RUIS PARA DECISÕES RUINS.

# Importar o comando pandas para dados.

import pandas as pd

dados = {
    "mes" : ["Jan", "Fev", "Mar", "Abr", "Mai", "Jun", "Jul", "Ago", "Sete", "Out", "Nov", "Dez"],
    "vendas_none": [1000, 2000, 5000, None, 6000, 7000, 8000, 9000, 10000, 11000, None, 12000],
    "vendas": [1000, 2000, 5000, 0, 6000, 7000, 8000, 9000, 10000, 11000, 0, 12000],
    "regiao": ["Sul", "Nordeste", "Sul", "Sudeste","Sul","Sul","Sul","Norte","Sul","Sul","Sudeste","Sul",]
}

df_dados = pd.DataFrame(dados)
# read_csv - lê csv
# read_json - lê json
# read_excel - lê excel
# DataFrame - lê dados informados no código

media_vendas = df_dados["vendas"].mean()
print(media_vendas, "correto")
media_vendas = df_dados["vendas_none"].mean()
print(media_vendas, "none")

print(df_dados["vendas"][3])
```

```python
# Solução para o caso none.
# Para corrigir um campo com erro de informação usamos o comando - fillna.
df_dados['vendas_corrigido'] = df_dados['vendas_none'].fillna(0)
display(df_dados)
```

# Gráfico com valores de vendas corrigidos.

```python
# Importar o comando matplotlib para gráficos.
# Importar o comando matplitlib = potly (deixar a visualização do gráfico mais bonita).

import matplotlib.pyplot as plt

#plt.bar(x,y)
plt.bar(df_dados['mes'],df_dados['vendas_corrigido'])
plt.show()
```

# Gráfico de Vendas Mensais em R$

```python
plt.bar(df_dados['mes'],df_dados['vendas_corrigido'])
plt.title('Vendas Mensais em R$')
plt.xlabel('Mês')
plt.ylabel('Valores em R$')
plt.grid()
plt.show()
```

# Corrigindo métricas erradas:


```python
# Métricas erradas:

dados2 = {
    "mes" : ["jan", "fev", "mar", "abr"],
    "vendas": [1000,1200,1100,1300]
}
df_dados2 = pd.DataFrame(dados2)
crescimento_vendas = df_dados2['vendas'].pct_change().mean()
print(crescimento_vendas)

# Corrigindo o valor:
print(crescimento_vendas*100,"%")

# Corrigindo casas decimais
# f'{variavel : Quantidade de casas decimais f}%'}
print(f'{crescimento_vendas*100:.2f}%')

#ou
print(f'{crescimento_vendas:.2%}')
```

# Gráfico de Linhas.

```python
plt.plot(df_dados2['mes'], df_dados2['vendas'])
plt.plot(df_dados2['mes'], df_dados2['vendas'].rolling(2).mean())
plt.plot(df_dados2['mes'], df_dados2['vendas'].rolling(3).mean())
plt.show()
```

# Gráfico de Barras.

```python
import matplotlib.pyplot as plt

# Calculate the mean of 'vendas_corrigido'
media_vendas_corrigido = df_dados['vendas_corrigido'].mean()

# Criar bar blot
plt.bar(df_dados['mes'], df_dados['vendas_corrigido'])

# Adicionando linha da media
plt.axhline(media_vendas_corrigido, color='red', linestyle='--', label=f'Média: {media_vendas_corrigido:.2f}')

plt.title('Vendas Mensais com Linha de Média em R$')
plt.xlabel('Mês')
plt.ylabel('Valores em R$')
plt.grid(True)
plt.legend() # Show the label for the mean line
plt.show()
```

# Gráfico de Linha.

```python
media_2 = df_dados2['vendas'].rolling(2).mean()
media_3 = df_dados2['vendas'].rolling(3).mean()

plt.plot(df_dados2['mes'], df_dados2['vendas'], label='Vendas')
plt.plot(df_dados2['mes'], media_2, label='Média 2 meses')
plt.plot(df_dados2['mes'], media_3, label='Média 3 meses')
plt.legend()
plt.show()
```
