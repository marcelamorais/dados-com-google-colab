```python
# Carregar os dados em formato JSON.
df_customers = pd.read_json('/content/customers_data.json')
df_products = pd.read_json('/content/products_data.json')
df_sales = pd.read_json('/content/sales_data.json')
```

```python
# Pré-diagnóstico:
# O comando info() mostra o nome das colunas, a quantidade de dados por coluna e o tipo de dado.
# O comando head() exibe os primeiros 5 linhas de dados.
df_customers.info()
df_customers.head()
```

```python
# O comando info() mostra o nome das colunas, a quantidade de dados por coluna e o tipo de dado.
# O comando head() exibe os primeiros 5 linhas de dados.
df_products.info()
df_products.head()
```

```python
# O comando info() mostra o nome das colunas, a quantidade de dados por coluna e o tipo de dado.
# O comando head() exibe os primeiros 5 linhas de dados.
df_sales.info()
df_sales.head()
```

```python
# Limpar e padronizar os dados - Costumers.

# Calcular a média e idade ausente:
media_idade = df_customers['age'].mean()

# Para substituir um campo devemos:
# 1 - Informar qual o campo será substituido.
# 2 - Informar que ele vai receber ele mesmo com a condição.
df_customers['age']= df_customers['age'].fillna(media_idade)

# Data Inválida:
df_customers['signup_date'] = pd.to_datetime(df_customers['signup_date'], errors= "coerce")

# Padronizar Gênero:
# .replace() = ("o que era", "o que vai ser")
# .str.upper() = (para deixar tudo maiúsculo).
df_customers['gender'] = df_customers['gender'].replace("", pd.NA).str.upper()
display(df_customers)
```

```python
# Limpar e padronizar os dados - Products.

# Trocar o basecost vazio pela média dos basecosts:
media_basecost = df_products['base_cost'].mean()
df_products['base_cost'] = df_products['base_cost'].fillna(media_basecost)
display(df_products)
```

```python
# Limpar e padronizar os dados - Sales.

# O sales, vendas está incorreto.
df_sales['sales'] = pd.to_numeric(df_sales['sales'], errors= "coerce")

# Vendas < 0.
# Localizar tudo que é menor que zero.
# 1 - Chama a base de dados com o atributo .loc
# 2 - O campo que eu quero.
# 3 - Informa a condição e o campo.
# 4 - O que vai substituir.

df_sales.loc[df_sales['sales']<=0, 'sales'] = pd.NA

# Converter as datas, que estão com diferentes padronizações.
df_sales ['date'] = pd.to_datetime(df_sales['date'], errors= "coerce")
print('Quantidade de dados truncados/errados.')
print(df_sales['date'].isna().sum())

# Corrigir a região:
# Para alterar um valor vazio que pode vir de diferentes maneiras (none"")
# podemos usar o comando .str.strip().str.title() para escrever o título que está la
# se era vazio quer dizer que não tem título, ou seja, mostra o texto NONE.
df_sales['region'] = df_sales['region'].str.strip().str.title()

# Corrigir o customer_type:
df_sales['customer_type'] = df_sales['customer_type'].replace("",pd.NA)
display(df_sales)
```

```python
# ATENCÃO - AQUI DADOS SERÃO EXCLUÍDOS/REMOVIDOS
# O comando dropna(subset=["CAMPOS A SEREM DETELATADOS"])

df_sales_limpos = df_sales.dropna(subset=['sales','date'])

# TABELAS DE CORRELAÇÕES - TABELA DINÂMICA DO EXCEL.
# Criar um campo de correlação entre os consumidores.
mapeamento_clientes = {
    "Novo": "C001",
    "Recorrente":"C002",
    "Corporativo":"C003"
}

# Vamos colocar o novo campo:
df_sales_limpos['customer_id'] = df_sales_limpos['customer_type'].map(mapeamento_clientes)
```

```python
# Visualizar colunas.
df_sales_limpos.columns
```

```python
df_consolidado = df_sales_limpos.merge(df_customers, on = 'customer_id')
```

```python
display(df_consolidado)
```

```python
df_consolidado = df_consolidado.rename(columns={"region_x":"region_sales"})
df_consolidado = df_consolidado.rename(columns={"region_y":"region_customer"})

# ATENCÃO - AQUI DADOS DE UMA DAS REGIÕES SERÃO EXCLUÍDOS/REMOVIDOS.
# df_consolidado = df_consolidado.dropna(columns=['region_y'])
display(df_consolidado)
```

```python
# Quantidade de vendas por região.

pd.pivot_table(
    # 1 - Qual é a base de dados.
    df_consolidado,
    # 2 - Quais são os valores.
    values="quantity",
    # 3 - Quais os index (legendas).
    index = 'region_sales',
    # Função de agregação.
    aggfunc="sum"
)
```

```python
# Ticket Médio por região.
# Correlacione o valor de venda por região.
pd.pivot_table(
    df_consolidado,
    values=['sales', 'cost'],
    index = "region_sales",
    aggfunc="mean"
)
```
