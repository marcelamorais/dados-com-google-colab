```python
# Importar o comando pandas para dados.
# Importar o comando numpy para matemática.
# Importar o comando matplotlib para gráficos.
# Importar o comando matplitlib = potly (deixar a visualização do gráfico mais bonita).
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Gerar um gráfico de barras.

```python
# Diagrama de Dispersão.
# Dados do DataFrame.
x= [10,20,30,40,50]
y= [15,25,35,45,55]

# Criar a visualização de dados:
plt.scatter(x,y, color = 'blue', label = "Visualização de Dados")
plt.title("Vendas no Tempo", fontsize =14)
plt.xlabel("Tempo", fontsize=11)
plt.ylabel("Vendas",fontsize = 11)
plt.legend()
plt.grid(True)

# Exibir a visualização de dados:
plt.show()
```

# Gerar gráfico de barras por categorias.

```python
# Gerar um gráfico de barras por categorias.
# Dados do DataFrame.
categoria = ['A','B','C','D','E'] #string
valores = ['10','40','30','65','8'] # int
plt.bar(categoria,valores, color = 'black', edgecolor = "yellow")
plt.title('Gráfico de barras por categorias', fontsize = 14)
plt.xlabel('CATEGORIAS', fontsize=11)
plt.ylabel('VALORES',fontsize = 11)
plt.grid(axis='y', linestyle = '--', alpha = 0.7)
```

```python
# Dados do DataFrame.
x= [10,20,50,60,70]
y= [15,65,35,45,85]

# Criar a visualização:
plt.plot(x,y, color = 'blue',label = 'Vendas Mensais', linestyle= '-')
plt.title("Acompanhamento de Vendas", fontsize =14)
plt.xlabel("Tempo", fontsize=11)
plt.ylabel("Vendas",fontsize = 11)
#plt.legend()
plt.grid(True)

# Exibir a visualização:
plt.show()
```

```python
# Montar bloxplot:
vendas_mensais = [300, 380, 360, 400, 450, 450, 500]

# O comando plt.figure pode ser utilizado pra box plot e também gráfico de pizza.
# O comando plt.pie pode ser utilizado para gráfico de setores usa "em baixo".
# O comando plt.boxplot pode ser utilizado para boxplot.
plt.figure(figsize = (8, 6))
plt.boxplot(vendas_mensais, patch_artist=True,
            boxprops=dict(facecolor='skyblue', color='black'), medianprops=dict(color='red'))
```
